# kaggle fraud detection

```python
import numpy as np
import matplotlib.pyplot as plt

preditores = ['Time','V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11', 'V12', 'V13', 'V14', 'V15', 'V16' , 
              'V17', 'V18', 'V19', 'V20', 'V21' , 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']
frequencia = [[3, 2], [15, 17], [8, 11], [15, 16], [24, 20], [13, 15], [4, 0], 
              [13, 17], [4, 10], [9, 3], [24, 25], [15, 12], [20, 18], 
              [3, 3], [25, 25], [0, 0], [13, 15], [12, 18], [5, 3], 
              [3, 2], [1, 2], [1, 2], [2, 5], [3, 0], [0, 1], [0, 1], [2, 1], [3, 3], [5, 0], [5, 3]]

sorted_data = sorted(zip(preditores, frequencia), key=lambda x: sum(x[1]), reverse=True)[:10]

preditores = [item[0] for item in sorted_data]
frequencia = [item[1] for item in sorted_data]
frequencia = np.array(frequencia)

x = np.arange(len(preditores))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, frequencia[:,0], width, label='SHAP', color="#008bfb")
rects2 = ax.bar(x + width/2, frequencia[:,1], width, label='LIME', color="#ff0051")

ax.set_ylabel('Número de vezes que apareceu no top 10')
ax.set_xlabel('Preditor')
ax.set_xticks(x)
ax.set_yticks(np.arange(26))
ax.set_xticklabels(preditores, rotation=90)
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.legend()

plt.grid(axis='y', alpha=0.3)
plt.grid(axis='x', alpha=0.3)
plt.minorticks_on()
plt.grid(which='minor', axis='x', color='gray', linestyle=':', alpha=0.3)

plt.savefig("plot_number_of_times_feature_in_top_ten_kaggle_fraud_detection.png")
fig.tight_layout()
plt.show()
```
